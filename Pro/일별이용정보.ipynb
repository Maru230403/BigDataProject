```python
# 파일 경로
file_path = 'data/2021/공공자전거 이용정보(일별)_2112.csv'

# 날짜 파싱 함수 정의
def custom_date_parser(date_str):
    try:
        # 날짜 문자열을 파싱하여 datetime 객체로 반환
        return pd.to_datetime(date_str, format='%Y-%m-%d')  # 날짜의 형식에 따라 format을 변경해주세요
    except ValueError:
        # 파싱에 실패하면 에러를 방지하기 위해 NaT (Not a Time) 반환
        return pd.NaT

# 데이터프레임으로 파일 읽기 (날짜 파싱 함수 사용)
df = pd.read_csv(file_path, parse_dates=['대여일자'], date_parser=custom_date_parser)
# '대여일자'를 날짜형으로 변환
#df['대여일자'] = pd.to_datetime(df['대여일자'])
#df.head()
```

```python
# '대여일자'사용 주차로 나누는 함수
def get_week(x):
    return x.week

```

```python
# '대여일자'를 주차로 나누어 '주차' 열 생성
df['주차'] = df['대여일자'].apply(get_week)
df.head()
```

```python
# 주차별 대여소별 대여건수
df_week = df.groupby(['주차', '대여소번호', '대여소']).size().reset_index(name='대여건수')
df_week.head()
```

```python
# 기존 텍스트 파일 경로
existing_file_path = 'data/2021/2021.csv'
# 데이터프레임을 'data/2019.txt' 파일에 추가 저장
df_week.to_csv(existing_file_path, mode='a', header=False, index=False, sep='\t', encoding='utf-8')
```
