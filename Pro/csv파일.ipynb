```python
def get_lat_lng(address):
    url = f'https://dapi.kakao.com/v2/local/search/address.json'
    headers = {'Authorization': f'KakaoAK {KAKAO_API_KEY}'}
    params = {'query': address}

    response = requests.get(url, headers=headers, params=params)
    data = response.json()

    if data['documents']:
        latitude = data['documents'][0]['y']
        longitude = data['documents'][0]['x']
        return latitude, longitude
    else:
        return None, None

# 주소에서 위도와 경도 가져와서 데이터프레임에 추가
df['Latitude'], df['Longitude'] = zip(*df['대여소'].apply(get_lat_lng).apply(pd.Series))

# 결과를 새로운 CSV 파일로 저장
output_csv_path = '2301_data.csv'
df.to_csv(output_csv_path, index=False)

print(f'위도와 경도가 추가된 파일이 저장되었습니다: {output_csv_path}')
```
